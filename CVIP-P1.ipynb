# Bivariate Analysis

```python
df[['gname','nkill']].groupby(['gname'],axis=0).sum().drop('Unknown').sort_values('nkill',ascending=False).head(10).plot(kind='bar',color='indigo',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Number of killed people",fontsize=15)
plt.show()
```

```python
pd.crosstab(df['region'],df['attacktype1'])
```

```python
pd.crosstab(df['weaptype1_txt'],df['attacktype1_txt'])
```

```python
df['gname'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='red',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack count",fontsize=15)
plt.show()
```
